```python
#LOAD Silver Purchase


from pyspark.sql import functions as F

silver_path = "/Volumes/workspace/default/gst-fraud&anomaly_detection/Silver/silver_purchase_orders"

silver_df = spark.read.parquet(silver_path)

display(silver_df.limit(5))

silver_df.count()
```

```python
silver_df.printSchema()
```

```python
from pyspark.sql.functions import current_timestamp

gold_anomaly_flags = (
    silver_df
    .select(
    "po_id",
    "vendor_id",
    "po_date",
    "invoice_number",
    "invoice_date",
    
    "is_gstin_valid",
    "blacklisted",
    "rate_mismatch",
    "value_spike",
    "name_mismatch",
    "duplicate_invoice",
    "director_overlap_flag",
    "vendor_non_filer",
    "missing_ewb",
    "state_mismatch",
    "late_invoice",
    "billing_state",

    "rule_fail_count",
    "rule_score",

    "is_anomalous",
    "severity",
    "anomaly_code"
    )
    .withColumn("gold_created_ts", current_timestamp())
)


display(gold_anomaly_flags.limit(10))
```

```python
from pyspark.sql.functions import *

gold_vendor_risk = (
    silver_df
    .groupBy(
        "vendor_id",
        "trade_name",
        "gstin"
    )
    .agg(
        sum("total_amount").alias("total_po_value"),

        count("*").alias("total_po_count"),

        sum("is_anomalous").alias("anomaly_count"),

        (
            sum("is_anomalous") / count("*")
        ).alias("anomaly_rate"),

        avg("name_sim_score").alias("avg_name_sim_score"),

        avg("filing_compliance_rate").alias("filing_compliance_rate")
    )
)
```

```python

#Risk Tier
from pyspark.sql.functions import when

gold_vendor_risk = gold_vendor_risk.withColumn(
    "risk_tier",
    when(col("anomaly_rate") >= 0.20, "Critical")
    .when(col("anomaly_rate") >= 0.10, "High")
    .when(col("anomaly_rate") >= 0.05, "Medium")
    .otherwise("Low")
)
```

```python
from pyspark.sql.functions import max, min, percentile_approx

stats_df = gold_vendor_risk.agg(
    max("anomaly_rate").alias("highest_anomaly_rate"),
    min("anomaly_rate").alias("lowest_anomaly_rate"),
    avg("anomaly_rate").alias("avg_anomaly_rate"))

display(stats_df)
```

```python
gold_vendor_risk.groupBy("risk_tier").count().show()
```

###Dashboard severity

```python
gold_anomaly_flags = gold_anomaly_flags.withColumn(
    "severity_dashboard",
    F.when(F.col("rule_fail_count") >= 6, "Critical")
     .when(F.col("rule_fail_count") >= 4, "High")
     .when(F.col("rule_fail_count") >= 2, "Medium")
     .otherwise("Low")
)
```

##ML prediction Join

```python
rf_predictions = spark.table("rf_predictions")

xgb_predictions = spark.table("xgb_predictions")
```

```python
gold_anomaly_flags = (
    gold_anomaly_flags
    .join(
        rf_predictions.select(
            "po_id",
            "rf_prediction",
            "rf_probability"
        ),
        "po_id",
        "left"
    )
    .join(
        xgb_predictions.select(
            "po_id",
            "xgb_prediction",
            "xgb_probability"
        ),
        "po_id",
        "left"
    )
)
```
